### describe() returns 8 statistical properties for each attribute
* Count: how many values the attribute has
* Mean: the mean of all values of the attribute 
* Standard Deviation: the amount of variation values have. Low $\sigma$ means values are closer to the mean
* Minimum value
* 25<sup>th</sup> Percentile: the value under which 25% of the instances can be found
* 50<sup>th</sup> Percentile: the value under which 50% of the instances can be found (also called the median, is the value that splits the distribution in half)
* 75<sup>th</sup> Percentile: the value under which 75% of the instances can be found
* Maximum value

## Remove and report the 20 least useful features from the sample

```python
for i in range(len(df_sample.dtypes)):
    if df_sample.dtypes[i] == 'object':
        col_to_change = df_sample.columns[i]
        df_sample = df_sample.astype({col_to_change: "int"})
```

```python
X = df_sample.iloc[:, :-1]
y = df_sample.iloc[:, -1]
# run the feature selection
from sklearn.ensemble import RandomForestClassifier
from sklearn.feature_selection import RFE
model_tree = RandomForestClassifier(random_state=100, n_estimators=50)
sel_rfe_tree = RFE(estimator=model_tree, n_features_to_select=34)
X_train_rfe_tree = sel_rfe_tree.fit_transform(X, y)
print(sel_rfe_tree.get_support())
```

```python
results = sel_rfe_tree.get_support()
features_to_drop = []
for i in range(len(df_sample.columns)-1):
    if not results[i]:
        features_to_drop.append(df_sample.columns[i])

# print the useless features
print(features_to_drop)
# and drop them
df_sample_feature_selected = df_sample.drop(columns=features_to_drop)
print(df_sample_feature_selected)
```

## Time to transform the dataset

```python
cut_bins = [0, 0.2, 0.4, 0.6, 0.8, 1]

for i in range(len(df_sample.dtypes)):
    if df_sample.dtypes[i] == 'float64':
        col_name = df_sample.columns[i]
        cut_labels = ['Very Low'+col_name, 'Low'+col_name, 'Medium'+col_name, 'High'+col_name, 'Very High'+col_name]
        df_sample_feature_selected[col_name] = pd.cut(df_sample_feature_selected[col_name], bins=cut_bins, labels=cut_labels)
```

```python
classes = {1:'one',2:'two',3:'three',4:'four',5:'five',6:'six',7:'seven'}
collection = []
for index, sample in df_sample_feature_selected.iterrows():
    item = []
    for col in df_sample_feature_selected.columns:
        if sample[col] == 0:
            continue
        # populate the item if 1
        if sample[col] == 1 and col != 'class':
            item.append(col)
            continue
        if col == 'class':
            item.append(classes[ sample[col] ])
            continue

        item.append( str(sample[col]) ) 
    collection.append(item)

print(collection)
```

```python
from apyori import apriori
# you should add the other attributes as well (bin the numeric ones)
association_rules = apriori(collection, min_support=0.1, min_confidence=0.7, min_lift=3, min_length=2)
association_results = list(association_rules)
```
